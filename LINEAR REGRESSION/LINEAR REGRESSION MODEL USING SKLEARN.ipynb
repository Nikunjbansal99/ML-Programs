```python
from sklearn import linear_model
from sklearn import metrics
from sklearn import model_selection
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(R'headbrain.csv')
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(
    np.array(df['Head Size(cm^3)']).reshape(-1,1),
    np.array(df['Brain Weight(grams)']).reshape(-1,1),
    test_size=0.2)
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
print('Mean squared error: ', metrics.mean_squared_error(
    Y_test, model.predict(X_test)))
x = np.linspace(2500,5000,10000)
```

```python
plt.scatter(np.array(df['Head Size(cm^3)']).reshape(-1,1),
            np.array(df['Brain Weight(grams)']).reshape(-1,1), c='r')
plt.plot(x, model.predict(x.reshape(-1,1)))
plt.title('Overall')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.show()
```

```python
def predict(X_new):
    Y_new = model.predict(X_new)
    return Y_new
```

```python
print('ENTER HEAD SIZE FOR BRAIN WEIGHT PREDICTION:')
X_new = float(input())
Z = float(predict(X_new))
print('PREDICTED VALUE OF BRAIN WEIGHT FOR GIVEN HEAD SIZE IS:',Z)
```
